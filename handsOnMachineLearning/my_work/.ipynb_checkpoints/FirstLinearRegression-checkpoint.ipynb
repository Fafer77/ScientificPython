```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
pokemon_dt = pd.read_csv('./Pokemon.csv')
X = pokemon_dt['Attack']
Y = pokemon_dt['Defense']
```

```python
pokemon_dt.plot(kind='scatter', x='Attack', y='Defense')
plt.show()
```

```python
model = LinearRegression()
model.fit(X, Y)
print(model.predict(112))
```
