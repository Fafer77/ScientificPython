# NOTES
Covaraince itself doesn't tell us much, but it is very useluf in correlation. As correlation normalizes covariance from -1 to 1 range so that it is meaningful. Covariance is mathematical generalization of variance.
Covariance matrix - when we take correlation of X and Y what we really get is matrix 2x2 where upper left corner is correlation X with X, bottom right corner is correlation Y with Y and right upper corner (left bottom corner) is correlation X with Y (Y with X - the same)

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
aaplData = yf.download('AAPL', '2021-01-01', '2024-06-06');
jpmData = yf.download('JPM', '2021-01-01', '2024-06-06');
aaplData = aaplData['Close']
jpmData = jpmData['Close']

print('Covariance of jpm and apple: \n' + str(np.cov(jpmData, aaplData)))
print('Correlation of jpm and apple: \n' + str(np.corrcoef(jpmData, aaplData)))
```

```python
mutual_fund_data = pd.read_csv('mutual_fund_data.csv')
expense = mutual_fund_data['Annual Expense Ratio'].values
sharpe = mutual_fund_data['Three Year Sharpe Ratio'].values

plt.scatter(expense, sharpe)
plt.xlabel('Expense Ratio')
plt.ylabel('Sharpe Ratio')

r_S = stats.spearmanr(expense, sharpe)
print('Spearman Rank Coefficient: ', r_S[0])
print('p-value: ', r_S[1])
```
