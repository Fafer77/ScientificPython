```python
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import datetime
```

```python
start = datetime.datetime(2022, 6, 6)
end = datetime.datetime(2024, 6, 6)
```

```python
cpi_data = pdr.get_data_fred('CPIAUCSL', start, end)
cpi_data
```

```python
cpi_data.plot(title="US CPI for All Urbans COnsumres", figsize=(10, 5))
```

```python
cpi_data['YoY'] = cpi_data['CPIAUCSL'].pct_change(periods=12)[1:] * 100
```

```python
cpi_data.pct_change()
```

```python
cpi_data['YoY'].plot(title='Year-over-Year CPI Change', figsize=(10, 5))
plt.xlabel('Date')
plt.ylabel('Percentage Change (%)')
plt.grid(True)
plt.show()
```
