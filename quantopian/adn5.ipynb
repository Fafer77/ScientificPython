```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the symbols and the time period
symbols = ['AAPL', 'MSFT']
start_date = '2024-01-01'
end_date = '2024-06-21'

# Download data
data = yf.download(symbols, start=start_date, end=end_date)

# Extract MSFT close prices
msft_close_prices = data['Close']['MSFT']

# Create a histogram with customized settings
plt.figure(figsize=(10, 6))
plt.hist(msft_close_prices, bins=20, alpha=0.75, color='blue', edgecolor='black')  # edgecolor for borders around bars
plt.xlabel('Close Price')
plt.ylabel('Number of Days Observed')
plt.title('Frequency Distribution of MSFT Prices, 2022-2024')
plt.grid(True)  # Enable grid lines

# Customize grid lines
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```

```python
jpm_close_prices = yf.download('JPM', start_date, end_date)
jpm_close_prices = jpm_close_prices['Close']
jpm_close_prices
```

```python
plt.scatter(msft_close_prices, jpm_close_prices)
plt.xlabel('MSFT')
plt.ylabel('JPM')
plt.title('Daily prices 2022-2024')
```

```python
plt.plot(msft_close_prices)
plt.plot(jpm_close_prices)
plt.legend(['MSFT', 'JPM'])
```
